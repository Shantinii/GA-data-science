### Review and Homework

For this homework, we will use the diamonds dataset to create the plots using Seaborn.


```python
# Make sure you have the necessary imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Read in the diamonds data set
diamonds = pd.read_csv('datasets/diamonds.csv')
diamonds.head()
```

Q1. Create a scatter plot using
- data: diamonds
- x-axis: depth
- y-axis: table
- color of markers: cut
- size of markers: price
- transparency: 0.5



```python
#Q1 Answer Using Seaborn
sns.scatterplot(data = diamonds, 
                x='depth', 
                y='table', 
                size ='price', 
                alpha = 0.5, 
                hue = 'cut')

plt.show()
```

Q2. Use facets to separate the plots by cut.


```python
#Q2 Using Seaborn

ax = sns.FacetGrid(diamonds, col='cut', hue ='cut')
ax.map(sns.scatterplot, 'depth', 'table')
ax.add_legend()
plt.show()
```

Q3. Draw Bar Charts to compare the mean prices by Cut and Clarity


```python
#Q3 Using Seaborn

ax = sns.barplot(data = diamonds, x = 'cut', y='price', 
                 hue ='clarity', errorbar=None)
ax.set_title("Mean Prices by Cut and Clarity")
ax.set_xlabel('')
plt.show()

```

```python
import numpy as np
fig, ax = plt.subplots(figsize=(20,8))
ax = sns.barplot(data = diamonds,
                 x = 'cut', 
                 y='price', 
                 hue='clarity',
                 ci= None, 
                 estimator=np.mean)
ax.set_title("Mean Prices by Cut and Clarity")
plt.show()
```

Q4. Generate the correlation matrix to check the correlation between the numeric variables in the diamonds dataset.

```python
# Q4 answer 
corr_mat_dia = diamonds.corr(numeric_only =True)
corr_mat_dia
```

Q5. Use the correlation matrix in a heatmap to determine the variables which are most highly correlated

```python
#Q5 Answer using seaborn

fig.ax = plt.subplots(figsize=(12,9))
ax = sns.heatmap(corr_mat_dia,
                annot=True,
                fmt='.2',
                cmap='coolwarm')
ax.set_title('Mean Price by Category')
plt.show()
```
